```python
import pandas as pd
import seaborn as sn

df = pd.read_csv("dataset_analise_sentimento.csv", sep=";")

df['data'] = pd. to_datetime(df['data'],format='%d/%m/%Y')

df_2018 = df.loc[pd.DatetimeIndex(df['data']).year == 2018]
df_2019 = df.loc[pd.DatetimeIndex(df['data']).year == 2019]
df_2020 = df.loc[pd.DatetimeIndex(df['data']).year == 2020]
df_2021 = df.loc[pd.DatetimeIndex(df['data']).year == 2021]

correlation_2018 = df_2018.corr()
correlation_2019 = df_2019.corr()
correlation_2020 = df_2020.corr()
correlation_2021 = df_2021.corr()

plot_2018 = sn.heatmap(correlation_2018, annot = True, fmt=".1f", linewidths=.6)
plot_2019 = sn.heatmap(correlation_2019, annot = True, fmt=".1f", linewidths=.6)
plot_2020 = sn.heatmap(correlation_2020, annot = True, fmt=".1f", linewidths=.6)
plot_2021 = sn.heatmap(correlation_2021, annot = True, fmt=".1f", linewidths=.6)
```
